```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf'
data = {
    'article_code': '',
    'article_type_list': '',
    'news_type_code': '',
    'show_what': '',
    'mediaCode': '',
    'lang': 'eng'
}

response = requests.post(url, data = data)
doc = BeautifulSoup(response.text)
doc
```

```python
# The article headline

headlines = doc.find_all('a', class_ = 'titlebet')
headlines
# class="titlebet"
for headline in headlines:
    print(headline.text)
```

```python
#links = doc.find_all('a', onclick = re.compile(r'^fn_showArticle\(\"AR.*'))

for onclick in headlines:
    print(onclick['onclick'])
    
    
# article_code: AR0126135
# article_type_list: 
# news_type_code: NT00
# show_what: L
# mediaCode: 
# lang: 
```

```python
df = pd.DataFrame(nk_list)
df
```

```python
df.to_csv('north_korea_news.csv', index = False)
```
