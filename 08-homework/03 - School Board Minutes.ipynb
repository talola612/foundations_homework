````python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
````

````python
response = requests.get('http://www.mineral.k12.nv.us/pages/School_Board_Minutes')
doc = BeautifulSoup(response.text)
doc
````

````python
items = doc.find('div', id = 'livesite-page-content-left').find_all('a', href = True, class_ = False)
items
````

````python
info_list = []

for item in items:
    if 'style' in str(item):
        info_dict = {}
        info_dict['date'] = pd.to_datetime(item.text)
        info_dict['url'] = item['href']
        info_list.append(info_dict)

df = pd.DataFrame(info_list)

````

````python
df
````

````python
df.to_csv('minutes.csv', index = False)
````
