```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
import seaborn as sns
```

```python
gender_df = pd.read_csv('data/gender-by-quarter.csv')
gender_df.head()
```

```python
gender_df['date'] = pd.to_datetime(gender_df['quarter'])
gender_df.head()
```

```python
plt.figure(figsize=(14,8))
sns.set_context(rc={"lines.linewidth": 6})

ax = sns.lineplot(data = gender_df, 
                  x = 'date', 
                  y = 'pct_female', 
                  hue = 'toa_group', 
                  legend = False,
                  marker = 'o',
                  markersize = 10,
                  palette = ['black', 'red'],
                  markeredgecolor = 'black',
                  markerfacecolor = 'white')

ax.set_yticks([0, 60], 10)
ax.set_ylabel('')
ax.set_xlabel('')
plt.xticks(rotation = 'vertical')

import matplotlib.dates as dates
ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%Y'))



ax.grid(axis = 'y', which = 'major', linestyle = '--')
ax.axvspan(dates.num2date(729391), 
           dates.num2date(730487), 
            facecolor='grey', alpha=0.2)

ax.axvspan(dates.num2date(733409), 
           dates.num2date(736331), 
            facecolor='grey', alpha=0.2)


plt.tight_layout()
plt.savefig("output.pdf")
```
