```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data/bank-profits.csv')
```

```python
df.head()
```

```python
labels = ['Asia Pacific', 'Middle East and Africa', 'Latin America', 'Central and Eastern Europe', 'Western Europe']
df.plot(x= 'region', y = 'year-2007', kind = 'pie')

plt.tight_layout()
plt.savefig("output.pdf")
```

```python
df.plot(x= 'region', y = 'year-2011', kind = 'pie')

plt.tight_layout()
plt.savefig("output1.pdf")
```

```python
fig, ax = plt.subplots(figsize = (9, 5))
ax.hlines(xmin = df.sort_values('year-2011', ascending = True).reset_index()['year-2007'], 
          xmax = df.sort_values('year-2011', ascending = True).reset_index()['year-2011'], 
          y = df.sort_values('year-2011', ascending = True).reset_index()['region'])
ax.scatter(x = df['year-2011'], y = df['region'])
ax.scatter(x = df['year-2007'], y = df['region'], color = 'skyblue')

plt.tight_layout()
plt.savefig("output.pdf")
```
