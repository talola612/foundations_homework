```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['pdf.fonttype'] = 42
%matplotlib inline
```

```python
poll_df = pd.read_csv('data/generic_poll_lead.csv') 
```

```python
poll_df.head()
```

```python
def build_colors(row):
    if row['in_power'] == 'yes':
        return 'gold'
    elif row['in_power'] == 'no':
        return 'lightgrey'

colors = poll_df.reset_index().apply(build_colors, axis=1)
colors

ax = poll_df.sort_values('year').plot(x = 'year', y = 'lead', kind = 'barh', figsize = (10, 9), color = colors, legend = False)

ax.set_title('2018 Doesnâ€™t Look Quite Like Other Election Cycles')

ax.legend(frameon = False,
          ncol=5, 
          loc='upper center',
          bbox_to_anchor=(0.5, 1.0),
          bbox_transform=plt.gcf().transFigure)
ax.set_ylabel('')

plt.tight_layout()
plt.savefig("nyt_poll.pdf")
```

```python
ss_df = pd.read_csv('data/social-spending.csv')
ss_df.head()

ss_df['ss_d'] = ss_df['Social Spending'].str.replace('%', '').astype(float)/100
```

```python
ss_df

ax = ss_df.sort_values('ss_d').plot(kind = 'barh', 
                                    figsize = (6, 20), 
                                    x = 'Country', 
                                    legend = False, 
                                    color = 'lightblue')
ax.set_ylabel('')


plt.tight_layout()
plt.savefig("nyt_ss.pdf")
```

```python
df = pd.read_csv('data/cook_pvi.csv')
```

```python
df['pct_undefeated'] = 100 - df.pct_defeated
df.head()
```

```python
ax = df.sort_values('pct_defeated').plot(x = 'cook_pvi', 
                                    y = ['pct_defeated', 'pct_undefeated'], 
                                    kind = 'barh',
                                    stacked = True,
                                    legend = False,
                                    figsize = (9, 5))

ax.set_ylabel('')

plt.tight_layout()
plt.savefig('nyt_cook_pvi.pdf')
```
