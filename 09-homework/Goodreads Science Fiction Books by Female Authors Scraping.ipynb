```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
books = doc.find_all('tr')
```

```python
rows = []

for book in books:
    row = {} 
    row['Rank'] = book.find(class_ = 'number').text.strip()
    row['Title'] = book.find(class_ = 'bookTitle').text.strip()
    row['Author'] = book.find(class_ = 'authorName').text.strip()
    row['Rating'] = book.find(class_ = 'minirating').text.strip()
    row['Score'] = book.find(class_ = 'smallText uitext').a.text.strip()
    row['Votes'] = book.find(class_ = 'smallText uitext').a.find_next_sibling('a').text.strip()
    rows.append(row)
rows
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df['Series'] = df.Title.str.extract('.*\((.*),? #.+\)', expand = False).str.replace(',', '')
df['Number in series'] = df.Title.str.extract('\(.*#(\d+)\)')
df['Title'] = df.Title.str.replace(' \(.*\)', '')
df['Number of ratings'] = df.Rating.str.replace(',', '').str.extract('.+ — (\d+) ratings')
df['Rating'] = df.Rating.str.replace(' avg rating — .* ratings', '')
df['Score'] = df.Score.str.replace('score: ', '').str.replace(',', '')
```

```python
df = df[['Rank', 'Title', 'Series', 'Number in series', 'Author', 'Rating', 'Number of ratings', 'Score', 'Votes']]
```

```python
df.to_csv('Goodreads_SciFi_Books_by_Female.csv', index = False)
```
