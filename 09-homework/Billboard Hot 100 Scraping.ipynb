```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
doc
```

```python
chart = doc.find_all(class_ = "chart-list-item")
```

```python
rows = []
for info in chart:
    row = {}
    row['song'] = info.find(class_ = 'chart-list-item__title-text').text.strip()
    row['artist'] = info.find(class_ = 'chart-list-item__artist').text.strip()        
    row['rank'] = info.find(class_ = 'chart-list-item__rank').text.strip()
    rows.append(row)

rows
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.to_csv('Billboard100.csv', index = False)
```
